# ü§ñ Predi√ß√£o com Machine Learning (Rede Neural MLP)

Este projeto utiliza **Machine Learning com Rede Neural Artificial (MLPClassifier)** para prever a *faixa de pre√ßo de celulares* a partir de vari√°veis t√©cnicas presentes no dataset.

O fluxo consiste em:
- üì• **Carregar os dados**  
- üìê **Normalizar as vari√°veis num√©ricas**  
- üîÄ **Dividir treino e teste**  
- üß† **Treinar uma Rede Neural (MLP)** com m√∫ltiplas camadas ocultas  
- üîÆ **Gerar previs√µes**  
- üìä **Calcular m√©tricas e avaliar o modelo**  
- üìÅ **Preparar DataFrames finais para visualiza√ß√£o no Power BI**

O resultado final inclui:
- Previs√µes da rede neural  
- Performance do modelo (acur√°cia, relat√≥rio de classifica√ß√£o e matriz de confus√£o)  
- Tabelas prontas para consumo no Power BI  

Uma solu√ß√£o completa de **ML + Rede Neural + BI**. üöÄ


### üì¶IMPORTs

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

### üóÇÔ∏èCarregando Base de Dados

```python
# Carregando Dados
caminho_arquivo = r'<CAMINHO PARA O csv>'
df_mobile_price = pd.read_csv(caminho_arquivo, sep=",")
```

### üîÑDividindo a base da Coluna de predi√ß√£o [ Y ]

```python
x_mobile_price = df_mobile_price.drop("price_range", axis=1)
y_mobile_price = df_mobile_price["price_range"]
```

### üóÑÔ∏èGuardando DF original para reuso [ antes do scaler ]

```python
x_mobile_price_original = x_mobile_price.copy()
```

### üîßNormalizando escala das Colunas n√∫mericas

```python
scaler_mobile_price = StandardScaler()
x_mobile_price = scaler_mobile_price.fit_transform(x_mobile_price)
```

### üèãÔ∏è‚Äç‚ôÇÔ∏èBase de treino para Base de Teste

```python
x_mobile_price_treinamento, x_mobile_price_producao,\
y_mobile_price_treinamento, y_mobile_price_producao,\
x_mobile_price_original_trainamento, x_mobile_price_original_producao = train_test_split(x_mobile_price, y_mobile_price, x_mobile_price_original, test_size=0.2, random_state=0)
```

### üß†Rede Neural


```python
mlp_mobile_price = MLPClassifier(
    max_iter=1000,                    # N√∫mero m√°ximo de √©pocas (itera√ß√µes) para treinar a rede
    tol=0.000001,                     # Toler√¢ncia m√≠nima para parar (crit√©rio de converg√™ncia)
    solver='adam',                    # Otimizador usado para ajustar os pesos (Adam = padr√£o e r√°pido)
    hidden_layer_sizes=(5, 5, 5, 5),  # Estrutura das camadas ocultas (4 camadas com 5 neur√¥nios cada)
    activation='relu',                # Fun√ß√£o de ativa√ß√£o usada nos neur√¥nios (ReLU = r√°pida e eficiente)
    random_state=0                    # Semente para deixar os resultados reproduz√≠veis
)

mlp_mobile_price.fit(
    x_mobile_price_treinamento,     # Dados de entrada para treinar o modelo
    y_mobile_price_treinamento      # R√≥tulos (classes) reais de cada amostra
)
```

### üîÆPrevisoes

```python
previsoes = mlp_mobile_price.predict(x_mobile_price_producao)
```

### üìàM√©tricas

```python
accuracy_score(y_mobile_price_producao, previsoes)
```

### üìÑCriando DataFrames para ser consumido no BI

```python
# base de referencia
df_resultado = x_mobile_price_original_producao.copy()

# resultado da classifica√ß√£o
df_resultado['price_range_real'] = y_mobile_price_producao

# resultado da rede neural
df_resultado['price_range_predito'] = previsoes

# DF do classification report
df_classification_repost = pd.DataFrame(classification_report(y_mobile_price_producao, previsoes, output_dict=True)).transpose().reset_index().rename(columns={'index': 'Classe'}).round(2)

# Matriz de confus√£o
df_matriz_confusao = pd.DataFrame(confusion_matrix(y_mobile_price_producao, previsoes), 
                                  columns=['Predito_0', 'Predito_1', 'Predito_2', 'Predito_3'],
                                  index=['Real_0', 'Real_1', 'Real_2', 'Real_3']).reset_index().rename(columns={'index': 'Classe Real'})
```
